```
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.decomposition import PCA
```

```
recipes = pd.read_csv('https://raw.githubusercontent.com/NewTorPG/muffin-cupcake/master/recipes_muffins_cupcakes_scones.csv')
recipes
```

```
ingredients = recipes[['Flour', 'Milk', 'Sugar', 'Butter', 'Egg', 'Baking Powder', 'Vanilla', 'Salt']].values
ingredients
```

```
type_label = np.where(recipes['Type']=='Muffin', 0, (np.where(recipes['Type']=='Cupcake', 1, 2)))
type_label
```

```
fold = list(StratifiedKFold(n_splits=5).split(ingredients, type_label))
pca = PCA(n_components=1)
ingriedients = pca.fit_transform(ingredients)
variance = pca.explained_variance_ratio_
model = svm.SVC(kernel='linear', C=0.01)
accuracy = cross_val_score(model, ingredients, type_label, cv=fold)
print('average accuracy=%g%%'%(100 * accuracy.mean()))
```
