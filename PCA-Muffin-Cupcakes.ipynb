```
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.decomposition import PCA
```

```
recipes = pd.read_csv('https://raw.githubusercontent.com/NewTorPG/muffin-cupcake/master/recipes_muffins_cupcakes.csv')
recipes
```

```
ingredients = recipes[['Flour', 'Milk', 'Sugar', 'Butter', 'Egg', 'Baking Powder', 'Vanilla', 'Salt']].values
ingredients
```

```
type_label = np.where(recipes['Type']=='Muffin', 0, 1)
type_label
```

```
fold = list(StratifiedKFold(n_splits=10).split(ingredients, type_label))
for i in range(1, 9):
  pca = PCA(n_components=i)
  ingriedients = pca.fit_transform(ingredients)
  variance = pca.explained_variance_ratio_
  model = svm.SVC(kernel='linear')
  accuracy = cross_val_score(model, ingredients, type_label, cv=fold)
  print('n_components=%g, average accuracy=%g%%'%(i, 100 * accuracy.mean()))
```
